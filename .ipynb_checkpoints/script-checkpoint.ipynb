```python
import os
import io
import sys
from contextlib import redirect_stdout
from yourPolygons import *
import difflib

directory = 'cpuTestFiles' 

if not os.path.exists(directory):
    print('There is no directory named {}, giving up...'.format(directory))
    sys.exit()

for filename in os.listdir(directory):
    if filename.endswith(".txt"):
        print(filename + " is testing...")
        
        try:
            polys = Polygons(os.path.join(directory, filename))
            str_io = io.StringIO()
            with redirect_stdout(str_io):
                polys.analyse()
            polys.display()
            output = str_io.getvalue()
        except Exception as e:
            output = "polygons.PolygonsError: " + str(e)

        output_filename = filename.replace('.txt', '_your_Answer.out')
        output_filepath = os.path.join(directory, output_filename)

        with open(output_filepath, 'w') as f:
            f.write(output)

print("Analysis complete, output files saved in", directory)

print()
print("Start comparing")
for filename in os.listdir(directory):
    if filename.endswith(".txt"):
        print(filename + " is comparing...")
        passed = True
        
        output_file = filename.replace(".txt", "_CPU_Answer.out")
        tex_file = filename.replace(".txt", "_CPU_TEX.tex")
        
        your_output_file = filename.replace(".txt", "_your_Answer.out")
        your_tex_file = filename.replace(".txt", ".tex")
        
        os.system('diff {} {}'.format(os.path.join(directory, output_file), os.path.join(directory, your_output_file)))
                  
        if "polys" in filename:
            os.system('diff {} {}'.format(os.path.join(directory, tex_file), os.path.join(directory, your_tex_file)))
        print()

```
